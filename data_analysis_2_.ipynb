```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
print(sns.get_dataset_names()) #gömülü kütüphanler
```

```python
bahsis_veri= sns.load_dataset('tips')
```

```python
print(bahsis_veri)
```

```python
print(type(bahsis_veri))
```

```python
#column sözlükte anahtar olarak eski kolon adları değer olarak yeni kolon anahtarları verilir.
#inplace=True ile değişikliklerin veri çerçevesinde kalıcı olması sağlanır.
bahsis_veri.rename(columns={'total_bill'            : 'fatura_toplami',
                               'tip'                : 'bahsis',
                               'sex'                : 'cinsiyet',
                               'smoker'             : 'sigara',
                               'day'                : 'gun',
                               'time'               : 'zaman',
                               'size'               : 'kisi_sayisi'
                              }, inplace=True)
```

```python
print(bahsis_veri.describe())
```

```python
sns.scatterplot(x='fatura_toplami', y='bahsis', data=bahsis_veri)
```

```python
sns.regplot(x='fatura_toplami', y='bahsis', line_kws={'color':'red'}, data=bahsis_veri)
```

```python
sns.set_style('darkgrid') #grfic style background color
sns.set(rc={'figure.figsize':(11,8)}) #grafik boyutu
sns.scatterplot(x='fatura_toplami', y='bahsis', data=bahsis_veri, hue='cinsiyet',color='blue',palette='coolwarm')
plt.xlabel('Fatura Toplamı') #xlabel name
plt.ylabel('Bahsis') #ylabel name
plt.title('Fatura Toplamı ve Bahsis') #title
plt.show()
```

```python
sns.scatterplot(x='fatura_toplami', y='bahsis', data=bahsis_veri, hue='sigara',style='cinsiyet',color='blue',palette='coolwarm')
plt.xlabel('Fatura Toplamı') #xlabel name
plt.ylabel('Bahsis') #ylabel name
plt.title('Fatura Toplamı ve Bahsis') #title
plt.show()
```

```python
sns.scatterplot(x='fatura_toplami', y='bahsis', size=("kisi_sayisi"), sizes=(20,200),data=bahsis_veri, hue='sigara',style='cinsiyet',color='blue',palette='coolwarm')
plt.xlabel('Fatura Toplamı') #xlabel name
plt.ylabel('Bahsis') #ylabel name
plt.title('Fatura Toplamı ve Bahsis') #title
plt.show()
```

```python
sns.scatterplot(x='fatura_toplami', y='bahsis', col="zaman",size=("kisi_sayisi"), sizes=(20,200),data=bahsis_veri,hue='sigara',style='cinsiyet',color='blue',palette='coolwarm')
plt.xlabel('Fatura Toplamı') #xlabel name
plt.ylabel('Bahsis') #ylabel name
plt.title('Fatura Toplamı ve Bahsis') #title
plt.show()
```

```python
sns.distplot(bahsis_veri['fatura_toplami']) #distplot
```

```python
sns.barplot(x='cinsiyet', y='bahsis', hue='cinsiyet',data=bahsis_veri) #barplot sns.barplot(data=df, x="island", y="body_mass_g", hue="sex")
```

```python
sns.catplot(x='cinsiyet', y='bahsis', hue='sigara',data=bahsis_veri,col='zaman',kind='bar') #catplot
```

```python
sns.catplot(x='cinsiyet', y='bahsis', hue='sigara',data=bahsis_veri,col='gun',kind='bar',) #catplot
```

```python
sns.heatmap(bahsis_veri.corr(), annot=True) #heatmap
```
