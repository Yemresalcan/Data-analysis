Question 2 - Find and visualize the IMDB values of the movies shot in the 'Documentary' genre between January 2019 and June 2020.

*Soru 2 - Ocak 2019 - Haziran 2020 tarihleri arasında 'Belgesel' türünde çekilen filmlerin IMDB değerlerini bulun ve görselleştirin.*

Question 3 - Which genre has the highest IMDB rating among movies shot in English?


---
*Soru 3 - İngilizce çekilen filmler arasında IMDB puanı en yüksek olan tür hangisidir?*


```python
df3 = df.loc[df['Language'].str.contains("English")]
df3 = df3.sort_values(['IMDB Score'], ascending=False).groupby('Genre').head(5)
df3 = df3.head(5)
df3
```

```python
fig = px.bar(df3, x='Genre', y='IMDB Score', text_auto=True, hover_name='Title',
             color='Title',
             color_discrete_sequence=['#D81F26'],template='plotly_dark',
             title="Ingilizce cekilen filmler icerisinde hangi tur en yuksek IMDB puanina sahiptir?", width=1400, height=700)
fig.update_layout(barmode='group')
fig.show()
```

Question 4 - What is the average 'runtime' of movies shot in 'Hindi'?


---

*Soru 4 - 'Hintçe' çekilen filmlerin ortalama 'çalışma süresi' nedir?*

```python
df4 = df.sort_values(['Runtime'], ascending=False).groupby('Language').head(35)
df4 = df4.loc[df['Language'].str.contains("Hindi")].head(35)
df4
```

```python
print("Filimerin ortalama süresi (Average duration of the filims):",df4['Runtime'].mean())
```

```python
fig = go.Figure(go.Indicator(
    mode = "gauge+number+delta",
    value = df4['Runtime'].mean(),
    domain = {'x': [0, 1], 'y': [0, 1]},
    title = {'text': "Avg. Runtime", 'font': {'size': 14}},
    delta = {'reference': df4['Runtime'].max(), 'increasing': {'color': "white"}},
    gauge = {
        'axis': {'range': [None, df4['Runtime'].max()], 'tickwidth': 1, 'tickcolor': "#D81F26"},
        'bar': {'color': "#D81F26"},
        'bgcolor': "white",
        'borderwidth': 2,
        'bordercolor': "gray",
}))

fig.update_layout(paper_bgcolor = "black", font = {'color': "white", 'family': "Arial"}, width=1400, height=700)
fig.show("")
```

---

Question 6 - Find the 3 most used languages in the movies in the data set

```python
df6 = pd.DataFrame(df)
df6 = df6.drop(columns=['Genre','Runtime','IMDB Score','Premiere','Title'])
#df6.head(50)

df6_keys = [df6['Language'].value_counts().index.tolist()[0],df6['Language'].value_counts().index.tolist()[1],
            df6['Language'].value_counts().index.tolist()[2]]
df6_values = [df6['Language'].value_counts()[0], df6['Language'].value_counts()[1], df6['Language'].value_counts()[2]]
df6_dictionary = {}
for i in range(len(df6_keys)):
    df6_dictionary[df6_keys[i]] = df6_values[i]

#df6_dictionary

df6_languages = list(df6_dictionary.keys())
df6_count = list(df6_dictionary.values())

fig = plt.figure(figsize = (15, 10))
colors = ['#D81F26', 'grey', 'grey']

plt.bar(df6_languages, df6_count, color = colors, width=0.6)
plt.xlabel("Language")
plt.ylabel("No. of movies")
plt.title("Number of Movies in the Dataset")
plt.show("png")
```
